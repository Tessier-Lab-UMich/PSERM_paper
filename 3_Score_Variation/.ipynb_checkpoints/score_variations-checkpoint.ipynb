```python
p1_clones = {s: 0 for s in p1_er.columns}
for s in p1_er.columns:
    p1_clones[s] = p1_er[s].dropna().index

p2_clones = {s: 0 for s in p2_er.columns}
for s in p2_er.columns:
    p2_clones[s] = p2_er[s].dropna().index

p3_clones = {s: 0 for s in p3_er.columns}
for s in p3_er.columns:
    p3_clones[s] = p3_er[s].dropna().index

dfs = []
er_dfs = []
keys = []

for s, p1_clone_set in p1_clones.items():
    dfs.append(np.log10(p1_data.uncertainties.loc[p1_clone_set, s] / abs(p1_data.scores.loc[p1_clone_set, f'{s} PSERM Score'].values) * 100))
    er_dfs.append(np.log10(p1_std.loc[p1_clone_set, s] / abs(p1_er.loc[p1_clone_set, s].values) * 100))
    keys.append(f'{s}')

for s, p2_clone_set in p2_clones.items():
    dfs.append(np.log10(p2_data.uncertainties.loc[p2_clone_set, s] / abs(p2_data.scores.loc[p2_clone_set, f'{s} PSERM Score'].values) * 100))
    er_dfs.append(np.log10(p2_std.loc[p2_clone_set, s] / abs(p2_er.loc[p2_clone_set, s].values) * 100))
    keys.append(f'{s}')

for s, p3_clone_set in p3_clones.items():
    dfs.append(np.log10(p3_data.uncertainties.loc[p3_clone_set, s] / abs(p3_data.scores.loc[p3_clone_set, f'{s} PSERM Score'].values) * 100))
    er_dfs.append(np.log10(p3_std.loc[p3_clone_set, s] / abs(p3_er.loc[p3_clone_set, s].values) * 100))
    keys.append(f'{s}')

pserm_df = pd.concat(dfs, axis=1, keys=keys).stack(0)
er_df = pd.concat(er_dfs, axis=1, keys=keys).stack(0)

pserm_df = pserm_df.reset_index(level=1)
er_df = er_df.reset_index(level=1)
```
